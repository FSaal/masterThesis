```python
def angle1(v1, v2):
    dot = np.dot(v1, v2)
    ang = np.arccos(dot)
    return ang

def angle3(v1, v2):
    dot = np.dot(unit_vector(v1), unit_vector(v2))
    if dot > 1:
        ang = np.arccos(dot)
    else:
        ang = 0
    return ang

def angle2(v1, v2):
    dot = np.dot(unit_vector(v1), unit_vector(v2))
    ang = np.arccos(dot)
    return ang
```

```python
%%heat
import numpy as np
from tf.transformations import unit_vector

def angle1(v1, v2):
    dot = np.dot(v1, v2)
    ang = np.arccos(dot)
    return ang

def angle3(v1, v2):
    dot = np.dot(v1, v2)
    if dot > 1:
        ang = np.arccos(dot)
    else:
        ang = 0
    return ang

def angle2(v1, v2):
    dot = np.dot(unit_vector(v1), unit_vector(v2))
    ang = np.arccos(dot)
    return ang

v1 = [-0.005327377002686262, -0.0032079138327389956, 0.9999806880950928]
v2 = [-0.005279263947159052, -0.0031344741582870483, 0.999981164932251]

for i in range(1000):
    angle1(v1, v2)
    angle2(v1, v2)
    angle3(v1, v2)
```
